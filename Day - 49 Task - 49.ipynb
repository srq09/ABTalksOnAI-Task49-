```python
"""
Task Day-49
Spam Classifier (Logic Level)

Author: Your Name
Description:
A rule-based spam classifier that identifies spam messages
using keyword detection, suspicious links, excessive symbols,
and promotional content.
"""

import re


class SpamClassifier:

    def __init__(self):

        self.spam_keywords = [
            "win",
            "winner",
            "free",
            "offer",
            "cash",
            "money",
            "lottery",
            "urgent",
            "congratulations",
            "click",
            "claim",
            "reward",
            "prize",
            "limited",
            "discount",
            "buy now"
        ]

    def extract_features(self, message):

        message_lower = message.lower()

        features = {
            "keyword_count": 0,
            "has_link": False,
            "has_phone_number": False,
            "uppercase_ratio": 0,
            "special_char_count": 0
        }

        # Spam keywords
        for keyword in self.spam_keywords:
            if keyword in message_lower:
                features["keyword_count"] += 1

        # URL detection
        if re.search(r"http[s]?://|www\.", message_lower):
            features["has_link"] = True

        # Phone number detection
        if re.search(r"\d{10}", message):
            features["has_phone_number"] = True

        # Uppercase ratio
        total_chars = len(message)

        if total_chars > 0:
            uppercase_chars = sum(
                1 for ch in message if ch.isupper()
            )

            features["uppercase_ratio"] = (
                uppercase_chars / total_chars
            )

        # Special character count
        features["special_char_count"] = len(
            re.findall(r"[!$#@%^&*]", message)
        )

        return features

    def classify(self, message):

        features = self.extract_features(message)

        spam_score = 0

        # Rule 1
        spam_score += features["keyword_count"] * 2

        # Rule 2
        if features["has_link"]:
            spam_score += 3

        # Rule 3
        if features["has_phone_number"]:
            spam_score += 2

        # Rule 4
        if features["uppercase_ratio"] > 0.30:
            spam_score += 2

        # Rule 5
        if features["special_char_count"] >= 3:
            spam_score += 2

        label = "SPAM" if spam_score >= 5 else "NOT SPAM"

        return {
            "message": message,
            "spam_score": spam_score,
            "classification": label,
            "features": features
        }


# ----------------------------
# Testing
# ----------------------------

classifier = SpamClassifier()

sample_messages = [

    "Congratulations! You have WON a FREE lottery. Click now!",
    
    "Meeting scheduled tomorrow at 10 AM.",
    
    "LIMITED OFFER!!! Buy now and get 80% discount.",
    
    "Please submit the assignment before evening.",
    
    "Claim your cash reward at www.rewardoffer.com"
]

for msg in sample_messages:

    result = classifier.classify(msg)

    print("\n" + "=" * 60)
    print("Message:")
    print(result["message"])

    print("\nSpam Score:")
    print(result["spam_score"])

    print("\nClassification:")
    print(result["classification"])

    print("\nFeatures:")
    print(result["features"])
```
